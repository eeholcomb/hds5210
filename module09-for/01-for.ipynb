## #2 - Triangular Number Sequence

There is a special sequence of numbers known as a triangular sequence.  It is generated by starting with `1` and adding `2` to that. Then adding `3` to the result.  Then adding `4` to the result of that.  The sequence continues by incrementally adding `1` more than the previous turn.  The first few numbers are shown below:

`1, 3, 6, 10, 15, ...`

Fill in the function below to find the `n`th number in that sequence.  Consider the answer `1` above to be the result of calling `triangle(1)` and `3` to be the result of `triangle(2)`, etc.  If `n` is less than 1, return 0.

````python
def triangle(n):
    
    # YOUR CODE HERE
    increment = 1
    answer = 0
    if n < 1:
        return 0
    else:
        for num in range(n):
            answer = answer + increment
            increment = increment + 1
    
        return answer
````

````python
# Feel free to test your code here
triangle(6)
````

````python
assert(triangle(1) == 1)
assert(triangle(2) == 3)
assert(triangle(3) == 6)
assert(triangle(4) == 10)
````
