# Getting Data from the Internet Using `requests`

**Steps:**

1. Find the URL for the data you want
2. Use `requests` to fetch a response from the internet
3. Check for a valid response and parse the resulting response text

```python
import requests
import csv
```

```python
url = 'http://dhcs-chhsagency.opendata.arcgis.com/datasets/8e4f3a0c75b9424d888d11c1f949cc32_0.csv'
```

```python
response = requests.get(url)
```

```python
text = response.text
```

```python
from io import StringIO
```

```python
f = StringIO(text)
reader = csv.reader(f)
data = []

for row in reader:
    data.append(row)
```

```python
len(data)
```

```python
data[0:10]
```

```python
data[0].index('Total_payments')
```

```python
total=0
total_idx = data[0].index('Total_payments')
for record in data[1:]:
    total += int(record[total_idx])
```

```python
"CA hospitals have received ${:,.2f} in payments.".format(total)
```
