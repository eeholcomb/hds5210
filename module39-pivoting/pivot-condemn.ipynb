```python
import pandas as pd
```

```python
condemns = pd.read_csv('/data/condemn.txt', low_memory=False)
```

```python
condemns.columns
```

```python
# After you do a groupby, the columns you grouped by become the data frame row index rather than a column.
# You can use .reset_index() to make them columns again!

by_pricinct_status = condemns.groupby(['Precinct02','Status'])['AcctNum'].count().reset_index()
```

```python
by_pricinct_status
```

```python
by_pricinct_status.pivot(index='Precinct02', columns='Status', values='AcctNum')
```

```python
import numpy as np
condemns.pivot_table(
    values='AcctNum', 
    index='InspectDate', 
    columns='Status', 
    aggfunc=np.ma.count, 
    fill_value=0)
```
